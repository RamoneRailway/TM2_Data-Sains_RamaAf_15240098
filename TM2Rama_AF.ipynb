```python
#Nama: Rama Afriansyah
#NIM: 15240098
#Kelas: 15.3A.01

import pandas as pd
import numpy as np

data = {
    'nama': ['Anna', 'Budi', 'Cica', 'Dewi', 'Eka', 'Fajar', 'Gaga', 'Hadi', 'Indah', 'Jilan'],
    'alamat': [
        'Jl. Merdeka No. 10', 'Jl. Sudirman No. 5', 'Jl. Gatot Subroto No. 15', 
        'Jl. Rasuna Said No. 8', 'Jl. Kuningan No. 12', 'Jl. Kuningan No. 12', 
        'Jl. Cikini No. 3', 'Jl. Palmerah No. 7', 'Jl. Senayan No. 9', 
        'Jl. Kuningan No. 2'
    ],
    'tinggi': [175, 178, 180, 170, 185, 185, 178, 168, 170, 160]
}

df = pd.DataFrame(data)

print("=== Data Calon TNI ===")
print(df)

print("\n=== Statistik Dasar Tinggi Badan ===")
stat_dasar = df['tinggi'].describe()

rata_rata = df['tinggi'].mean()
tertinggi = df['tinggi'].max()
terendah = df['tinggi'].min()
median = df['tinggi'].median()

stat_dasar.index = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']
print(stat_dasar)
print(f"Name: tinggi, dtype: float64")

print("\n== Tambahan Statistik ==")
print(f"Rata-rata tinggi : {rata_rata:.2f} cm")
print(f"Tertinggi        : {tertinggi} cm")
print(f"Terendah         : {terendah} cm")
print(f"Median           : {median} cm")

print("\n=== Calon di atas rata-rata tinggi ===")
df_di_atas_rata = df[df['tinggi'] > rata_rata]

print(df_di_atas_rata[['nama', 'alamat', 'tinggi']].reset_index(drop=True))

print("\n=== Urutan dari Tertinggi ke Terendah ===")
df_sorted = df.sort_values(by='tinggi', ascending=False)

print(df_sorted[['nama', 'alamat', 'tinggi']].reset_index(drop=True))

print("\n=== Data dengan Kategori Tinggi ===")
bins = [0, 169, 179, np.inf] # Batas: (0-169, 170-179, 180-inf)
labels = ['Pendek', 'Sedang', 'Tinggi']

df['kategori'] = pd.cut(df['tinggi'], bins=bins, labels=labels, right=True)

print(df)

print("\n=== Jumlah Calon per Kategori ===")
jumlah_kategori = df['kategori'].value_counts().sort_index()

print(jumlah_kategori)
print(f"Name: count, dtype: int64")

print("\n=== Statistik Lanjutan ===")
std_deviasi = df['tinggi'].std()
variansi = df['tinggi'].var()

print(f"Standar Deviasi : {std_deviasi:.2f}")
print(f"Variansi        : {variansi:.2f}")

print("\n=== Calon dengan Tinggi Badan Tertinggi ===")
max_tinggi = df['tinggi'].max()

calon_tertinggi = df[df['tinggi'] == max_tinggi]

if not calon_tertinggi.empty:
    calon_tertinggi_fajar = calon_tertinggi[calon_tertinggi['nama'] == 'Fajar'].iloc[0]
    
    print(f"Nama   : {calon_tertinggi_fajar['nama']}")
    print(f"Alamat : {calon_tertinggi_fajar['alamat']}")
    print(f"Tinggi : {calon_tertinggi_fajar['tinggi']} cm")
```
